```python
# Day 12 of Advent of code 2020, part one;
#
# Ship navigation!
# I need a x, y updater which can follow the input directions.
# Ship starts facing east,
#
degreesToFacing = {0: "N", 90: "E", 180: "S", 270: "W"}
FacingToDegrees = {"N": 0, "E": 90, "S": 180, "W": 270}
directions = ["N","E","S","W"]
x1 = [0, 0, "E"]

# Def preparing prepares variables for moving the boat, returns newX, newY and currentFacing
def preparing(directions, currentX, currentY, currentFacing):
    Amount = int(directions[1:])
    instruction = directions[0]
    if instruction == "N" or instruction == "E" or instruction == "S" or instruction == "W":
        x = move(currentX, currentY, instruction, Amount, currentFacing)
        return x
    elif instruction == "R":
        degrees = FacingToDegrees[currentFacing] + Amount
        if degrees>=360:
            degrees -= 360
        elif degrees<0:
            degrees += 360
        currentFacing = degreesToFacing[degrees]
        return currentX, currentY, currentFacing
    elif instruction == "L":
        degrees = FacingToDegrees[currentFacing] - Amount
        if degrees>=360:
            degrees -= 360
        elif degrees<0:
            degrees += 360
        currentFacing = degreesToFacing[degrees]
        return currentX, currentY, currentFacing
    elif instruction == "F":
        x = move(currentX, currentY, currentFacing, Amount, currentFacing)
        return x

def move(currentX, currentY, instruction, moveAmount, currentFacing):
    print(f"Moving forward {moveAmount}, towards {instruction}")
    if instruction == "N":
        currentY += moveAmount
    if instruction == "E":
        currentX += moveAmount
    if instruction == "S":
        currentY -= moveAmount
    if instruction == "W":
        currentX -= moveAmount
    return currentX, currentY, currentFacing


# File handling
f = open("directions_day12.txt", "r")
line = f.readline()
while line:
    print(line)
    x1 = preparing(line, x1[0], x1[1], x1[2])
    print(f"Facing: {x1[2]}, X = {x1[0]}, Y = {x1[1]}")
    line = f.readline()

ending_x_1 = abs(int(x1[0]))
ending_y_1 = abs(int(x1[1]))
print(f"The result of challange 1 is {ending_x_1+ending_y_1}")
    
```
