```python
# Day 

f = open("trees_day3.txt", "r")
xPos = 0
bomen = 0
line = f.readline()
length = len(line)-1

while len(line)>0:
    if line[(xPos%length)] == "#":
        bomen +=1
    #print(f"{line}")
    #line2 = line[0:xPos%length] + "X" + line[xPos%length+1: ]
    #print(f"{line2}")
    #print(f"De numerieke positie is {xPos%length} (xPos%11), en {xPos%length-1} (xPos%11-1).")
    #print(f"Bij deze positie hoort letter {line[xPos%length]} (xPos%11), en {line[xPos%length-1]} (xPos%11-1)")
    #print(f"Er zijn {bomen} bomen!")
    #print("------------------------")
    xPos += 1
    print(bomen)
    line = f.readline()
print(bomen)
print("Bij (1 rechts, 1 beneden) raak je 68 bomen!")
print("Bij (3 rechts, 1 beneden) raak je 203 bomen!")
print("Bij (5 rechts, 1 beneden) raak je 78 bomen!")
print("Bij (7 rechts, 1 beneden) raak je 77 bomen!")
print("Bij (1 rechts, 2 beneden) raak je 40 bomen!")
uitkomst = 68*203*78*77*40
print(uitkomst)
```
