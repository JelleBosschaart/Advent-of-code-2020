```python
# Day 5 Challange one

#For example, consider just the first seven characters of FBFBBFFRLR:

#Start by considering the whole range, rows 0 through 127.
#F means to take the lower half, keeping rows 0 through 63.
#B means to take the upper half, keeping rows 32 through 63.
#F means to take the lower half, keeping rows 32 through 47.
#B means to take the upper half, keeping rows 40 through 47.
#B keeps rows 44 through 47.
#F keeps rows 44 through 45.
#The final F keeps the lower of the two, row 44.

total_rows = 127
total_seats = 7

import math

def seat_row(seating_binary):
    rowsLower = 0
    rowsUpper = total_rows
    seatsLower = 0
    seatsUpper = total_seats 
    
    
    for i in seating_binary:
        if i == "F":
            rowsUpper = math.floor((rowsUpper+rowsLower)/2)
        if i == "B":
            rowsLower = math.ceil((rowsUpper+rowsLower)/2)
        if i == "L":
            seatsUpper = math.floor((seatsUpper+seatsLower)/2)
        if i == "R":
            seatsLower = math.ceil((seatsUpper+seatsLower)/2)
    row = rowsLower
    seat = seatsLower
    seatID = (row*8) + seat
    return row, seat, seatID
    
f = open("boarding_passes_day5.txt", "r")
line = f.readline()

d = 0
while line:
    x = seat_row(line)
    print(line)
    print(x)
    if x[2] > d:
        d = x[2]
    line = f.readline()
print(d)
```
