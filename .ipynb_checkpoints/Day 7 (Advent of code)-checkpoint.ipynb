```python
f = open("bag_rules_day7.txt", "r")
line = f.readline()


def bag_identify(string):
    print(string) # debug
    amounts = []
    bags = []
    
    # Get the amounts of bags, in a list with integers.
    for i in string:
        if i.isdigit():
            amounts.append(int(i))
            
            u = string.find(i)
            string2 = string[u:]
            i = string2.find(i)
            comma = string2.find(",")
            if comma == -1:
                comma = string2.find(".")
            print(f"comma is {comma}")
            bags.append(string2[i+2:comma])
            
    
    contain = string.find("contain")
    original_bag = string[0:contain-6]
    bags_contained = string[contain+8:]
    
    # debug
    print(original_bag)
    print(amounts)
    print(bags)
    print(bags_contained)
    
    
while line:
    bag_identify(line)
    line = f.readline()
```
