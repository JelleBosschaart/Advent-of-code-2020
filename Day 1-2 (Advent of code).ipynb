```python
# Day 2 part 2
import re
total = 0
    
def find_values(string):
    bound = re.findall(r'\d+',string) # lowerBound = bound[0] and upperBound = bound[1]
    lMin = int(bound[0])
    lMax = int(bound[1])
    posLetter = string.find(":")-1 # test[letter] = letter in password
    letter = string[posLetter]
    password = string[string.find(":")+2:len(string)-1]
    return lMin, lMax, letter, password

def password_check(lMin, lMax, letter, password):
    checking = 0
    if password[lMin-1] == letter:
        checking += 1
    if password[lMax-1] == letter:
        checking += 1
    if checking == 1:
        return "true"
    else:
        return "false"

f = open("calender_day2.txt", "r")
line = (f.readline())
while len(line)>0:
    output = find_values(line)
    check = password_check(output[0], output[1], output[2], output[3])
    if check == "true":
        total += 1
    print(line)
    line = (f.readline())
print(total)
        
```
