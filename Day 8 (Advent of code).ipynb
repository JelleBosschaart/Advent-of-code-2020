```python
# Day 8, part two
# This is definitly not an efficient option, but it works.

accumulator = 0
database = []
memory = []

def switch(line):
    print(f"Switching: {line}")
    if "nop" in line:
        newline = "jmp "  + line[4] + line[5:]
    if "jmp" in line:
        newline = "nop " + line[4] + line[5:]
    if "acc" in line:
        newline = "acc " + line[4] + line[5:]
    print(f"Done! Switched to: {newline}")
    return newline
    
def working_lines(database, trie):
    i = 0
    accumulator = 0
    memory = []
    
    while i < len(database):
        
        line = database[i]
        if i in memory:
            database[trie] = switch(database[trie])
            return "false"
        memory.append(i)
        
        if "acc" in line:
            argument = int(line[5:])
            if "-" in line:
                accumulator -= argument
            elif "+" in line:
                accumulator += argument
            i += 1
    
        # The jump function
        if "jmp" in line:
            argument = int(line[5:])
            if "-" in line:
                i -= argument
            elif "+" in line:
                i += argument
        if "nop" in line:
            i += 1
            
    print("done!")
    print(f"the final accumilator was: {accumulator}")
    print(f"The final i was: {i}/{len(database)}")
    print(f"Chaning the second to last line in database, then trying again.")


# Line handling, creating lines database for followthru;
f = open("infinite_loop_day8.txt", "r")
line = f.readline()
while line:
    database.append(line[:-1])
    line = f.readline()

    
for i in range(len(database)):
    database[i] = switch(database[i])
    if working_lines(database, i) == "false":
        print(f"{i} didnt work..")
    else:
        solution = i
        exit()
        
        print(f"{i} DID WORK! WOOOOOOO")

#ds = working_lines(database)
#working_lines(ds)
print(f"the final accumilator is: {accumulator}")
```
